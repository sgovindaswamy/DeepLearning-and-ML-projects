```python
import pandas as pd
import yfinance as yf
import datetime

# Download the gold dataset from Yahoo Finance (5 years of data)
symbol = "GC=F"  # Symbol for gold futures
end_date = datetime.datetime.today().strftime('%Y-%m-%d')  # Today's date
start_date = (datetime.datetime.today() - pd.DateOffset(years=5)).strftime('%Y-%m-%d')  # 5 years ago

gold_data = yf.download(symbol, start=start_date, end=end_date)

# Calculate the 200-day moving average for the past 5 years
gold_data['200_MA'] = gold_data['Close'].rolling(window=200).mean()

# Function to calculate the weight based on how close the price is to the moving average
def calculate_weight(close, ma):
    if pd.isna(ma):  # Handle NaN values
        return None
    diff_percentage = abs(close - ma) / ma * 100
    if diff_percentage <= 25:
        return 1
    elif diff_percentage <= 50:
        return 0.75
    elif diff_percentage <= 75:
        return 0.5
    elif diff_percentage <= 100:
        return 0.25
    else:
        return 0

# Apply weight calculation row-wise
gold_data['Weight'] = gold_data.apply(
    lambda row: calculate_weight(row['Close'], row['200_MA']), axis=1
)

# Drop rows with NaN in 'Weight'
gold_data = gold_data.dropna(subset=['Weight'])

# Group by month to estimate units to buy each month
gold_data['Month'] = gold_data.index.to_period('M')

# Assuming a fixed monthly budget (e.g., $1000)
monthly_budget = 1000

# Calculate units to buy each month
monthly_units = gold_data.groupby('Month').apply(
    lambda x: (monthly_budget * x['Weight'].sum()) / x['Close'].mean()
)

# Print monthly units to buy
print("Monthly Units to Buy in 2025 (based on 200-day MA strategy):")
print(monthly_units)

# Assuming that you're interested in total units to buy over the year (2025)
total_units_2025 = monthly_units.sum()
print(f"\nTotal Units to Buy in 2025: {total_units_2025}")
```

```python
# Perform cumulative sum on 'Value_5th' column
gold_data['Cumulative_Value_5th'] = gold_data['Value_5th'].cumsum()

# Display the result
print(gold_data[['Value_5th', 'Cumulative_Value_5th']].dropna())

# Optional: Plot the cumulative values
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(gold_data['Cumulative_Value_5th'], label='Cumulative Value', color='blue', marker='o')
plt.title('Cumulative Value for Every 5th Row')
plt.xlabel('Date')
plt.ylabel('Cumulative Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a series of values starting from 100, incremented by 100, only for every 5th row
cash_in_values = np.arange(100, 100 * (len(gold_data) // 5 + 1), 100)

# Assign these values to every 5th row (starting from index 4)
gold_data['cash_in'] = np.nan  # Initialize with NaN
gold_data.iloc[4::5, gold_data.columns.get_loc('cash_in')] = cash_in_values

# Display the DataFrame to check the result
print(gold_data[['Value_5th', 'Cumulative_Value_5th', 'cash_in']].dropna())  # Drop NaNs for easier viewing
```

```python
# Initialize the new column with NaN values for 'returns'
gold_data['returns'] = np.nan

# Loop through every 5th row to calculate the 'returns'
for i in range(4, len(gold_data), 5):  # Start at the 5th row (index 4)
    # Fetch scalar values for cumulative_value, closing_price, and cash_in at every 5th row
    cumulative_value = gold_data['Cumulative_Value_5th'].iloc[i]
    
    # Extract the scalar value from the 'Closing_Price' Series (access the first element)
    closing_price = gold_data['Closing_Price'].iloc[i].values[0]
    
    cash_in = gold_data['cash_in'].iloc[i]
    
    # Debugging step: Print the values of cumulative_value, closing_price, and cash_in
    print(f"Index {i} - Cumulative Value: {cumulative_value}, Closing Price: {closing_price}, Cash In: {cash_in}")

    # Check if all scalar values are not null (handle each value separately)
    if pd.notnull(cumulative_value) and pd.notnull(closing_price) and pd.notnull(cash_in):
        # Perform the calculation: (Cumulative_Value_5th * Closing_Price) / cash_in
        returns = (cumulative_value * closing_price) / cash_in
        gold_data.at[gold_data.index[i], 'returns'] = returns

# Display the DataFrame to check the result
print(gold_data[['returns']].dropna())  # Drop NaNs for easier viewing

```
